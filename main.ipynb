```python
# main_project.ipynb
from car_rental import CarRental, Customer
from datetime import datetime

#Main function
def main():
    available_cars = {'Tata Tiago EV': 10, 'Tata Nexon EV': 8, 'MG Comet EV': 6}
    car_rental = CarRental(available_cars)
    customer = Customer()

    while True:
        print("\nOptions:")
        print("1. Display Available Cars")
        print("2. Rent a Car")
        print("3. Return a Car")
        print("4. Exit")
        choice = input("Enter your choice: ")

        if choice == '1':
            car_rental.display_available_cars()
        elif choice == '2':
            car = input("Enter car name: ")
            num_cars = int(input("Enter number of cars: "))
            rent_time = datetime.now()
            rent_period = input("Enter rental period (hourly/daily/weekly): ")
            customer.request_rental(car_rental, car, num_cars, rent_time, rent_period)
        elif choice == '3':
            car = input("Enter car name: ")
            rent_time = datetime.now()
            customer.return_rental(car_rental, car, rent_time)
        elif choice == '4':
            print("Thank you for using our car rental service.")
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```
